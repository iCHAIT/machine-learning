```python
import numpy as np

%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('/Users/chaitanyagupta/Desktop/machine-learning/Linear_Regression/ex1/ex1data1.txt', delimiter=',')

x = data[:, 0]
y = data[:, 1]

try:
    m, n = x.shape[0], x.shape[1]
except IndexError:
    m, n = x.shape[0], 1
    

x0 = np.ones(m)

X = np.c_[x0, x]
```

### 2.1 Plotting Data

```python
plt.figure(figsize=(10,6))
plt.plot(x, y, 'rx', markersize=10)
# plt.grid(True)
plt.ylabel('Profit')
plt.xlabel('Population of City')
```

### 2.2 Gradient descent & Cost Function

```python
def J(theta, X, y):
    
    # Calculate hypothesis
    hypothesis = np.dot(X, theta)
    
    # Calculate loss
    loss = hypothesis - y
    
    # Error
    error = 1/(2*m) * np.sum(loss ** 2) 
    
    return error
    
    
    
l_rate = 0.01
theta = np.zeros(n+1)

iters = 1500

print("Learining Rate: ", l_rate)
print("Initial theta: ", theta)

# Initial Error with theta init to 0
Jn = J(theta, X, y)
print("Initial Error: ", Jn)

error_pts = []
    
for i in range(iters):
    
    Jn = J(theta, X, y)
    error_pts.append(Jn)

    theta = theta - l_rate * (1/m) * np.dot((np.dot(X, theta) - y), X)
    
    

print("Final Error: ", Jn)
print("No of Iterations taken for conv: ", iters)
print("Parametrs: ", theta)
```

### Plotting Convergence

```python
# Plot convergence of cost function

def plotConvergence(error_pts):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(error_pts)), error_pts, 'bo' )
    plt.title("Convergence of Cost Function")
    plt.xlabel("Iteration number")
    plt.ylabel("Cost function")
```

```python
plotConvergence(error_pts)
```

### Plotting hypothesis on training data

```python
#Plot the line on top of the data to ensure it looks correct
    
y_pts = []
    
for i in x:
    y_new = theta[0] + theta[1]*i
        
    y_pts.append(y_new)
        

def plotline(x, y, theta, y_pts):
    plt.figure(figsize=(10,6))
    
    # Plot training data
    plt.plot(x, y ,'rx',markersize=10,label='Training Data')
    
    # Plot hypothesis line on same graph
    plt.plot(x, y_pts, 'b-',label = 'Hypothesis: h(x) = %0.2f + %0.2fx'%(theta[0],theta[1]))
    plt.ylabel('Profit')
    plt.xlabel('Population of City')
    plt.legend()
        

    
    
plotline(x, y, theta, y_pts)
```
